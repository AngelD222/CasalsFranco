## Introducción
Este repositorio contiene la implementación estructurada y el estudio comparativo de diversos algoritmos clásicos de aprendizaje por refuerzo orientados a resolver el problema del **Bandido Multibrazo**. 

El desafío central es encontrar el equilibrio perfecto entre la **exploración** de nuevas alternativas y la **explotación** de las conocidas. A través de simulaciones de Monte Carlo, contrastamos el rendimiento de diversas políticas analizando métricas como la recompensa media, el rechazo acumulado (regret) y el Error Cuadrático Medio (MSE) en las estimaciones de valor.
