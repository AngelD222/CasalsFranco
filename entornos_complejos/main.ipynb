## Introducción
A diferencia del problema del bandido de k-brazos, donde las recompensas dependen únicamente de la acción elegida, muchos problemas reales de aprendizaje por refuerzo se desarrollan en entornos complejos donde las decisiones dependen del estado actual del sistema y afectan a estados futuros. Este tipo de problemas se modela mediante **Procesos de Decisión de Markov (MDP)**.

En estos entornos, el agente interactúa con el entorno observando un estado, seleccionando una acción según su política, recibiendo una recompensa y transitando a un nuevo estado. El objetivo del agente es aprender una política que maximice la recompensa acumulada a lo largo del tiempo.

En esta parte del trabajo se estudian diferentes técnicas de aprendizaje por refuerzo en entornos de Gymnasium, incluyendo métodos tabulares como **Monte Carlo, SARSA y Q-Learning**, así como métodos con aproximación de funciones como **Deep Q-Learning y SARSA semi-gradiente**. El objetivo es analizar su comportamiento y comparar su rendimiento en distintos entornos.
